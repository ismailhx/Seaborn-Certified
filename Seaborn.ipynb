# **Seaborn**

### Graphs

Import Seaborn ( a library built on top of Matplotlib )

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
flights = sns.load_dataset("flights")
# loads a seaborn dataset which is located in the seaborn data on github 
# https://github.com/mwaskom/seaborn-data
flights.head(20)
```

Drawing a simple x y plot

```python
sns.lineplot(x = "year", y = "passengers", data = flights)
plt.show()
```

Styleing the line

```python
sns.lineplot(x = "year", y = "passengers", data = flights, color="red", linestyle="--", linewidth=2)
plt.show()
```

Styling a selected column 



```python
sns.lineplot(x = "year", y = "passengers", data = flights, hue = "month")
plt.figure(figsize=(16,7))
plt.show()
```

Here the month is styled - each month is given a different type of line

```python
sns.lineplot(x = "year", y = "passengers", data = flights, hue = "month", style = "month")
plt.show()
```

Adding markers using hue =

```python
sns.lineplot(x = "year", y = "passengers", data = flights, hue = "month", style = "month", markers = True)
plt.show()
```

## **ScatterPlots**

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris)
plt.show()
#as petal length increases so does sepal length
```

The colour of the dots should be based on the species using hue=

The shape of the markings is determined by Style

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris, hue = "species", style = "species")
plt.show()
```

Determining the hue on a numerical feature

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris, hue = "petal_length")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris, hue = "petal_length", style = "species", palette = "Blues_d")
plt.show()
```

## **Histograms, Distribution / distplots & Joint plots**

##Distribution Plots

```python
penguins = sns.load_dataset("penguins")
penguins
```

#### Shows the distribution of the flipper length column

#### This is a combination of a histogram and frequency curve

Histograms - make a number of bins 
- each bin is a container of values which is a category e.g 175-180 flipper mm size

Frequency distribution aka KDE measure the frequency of something

```python
sns.distplot(penguins.flipper_length_mm)
plt.show
```

A kde/frequency curve without the histogram:

```python
sns.distplot(penguins.flipper_length_mm, color = "orange", hist = False)
plt.show
```

A histogram without the frequency curve

```python
sns.distplot(penguins.flipper_length_mm, bins = 10, kde = False)
plt.show
```

Adding a different colour to the distplot

```python
sns.distplot(penguins.flipper_length_mm, color = "green")
plt.show
```

Plotting the distribution plot on the y axis

```python
sns.distplot(penguins.flipper_length_mm, vertical = True)
plt.show
```

## **Jointplots**

You have the histogram for flipper length and body mass

```python
sns.jointplot(x = "flipper_length_mm", y = "body_mass_g", data = penguins)
plt.show
```

Changing the colour of the histogram

```python
sns.jointplot(x = "flipper_length_mm", y = "body_mass_g", data = penguins, color = "olive")
plt.show
```

Adding a regression line throught the scatter plot and the histogram

```python
sns.jointplot(x = "flipper_length_mm", y = "body_mass_g", data = penguins, color = "olive", kind = "reg" )
plt.show
```

## **Working with boxplots**

```python
sns.boxplot(y ="body_mass_g", x ="sex" , data = penguins)
```

```python
sns.boxplot(y ="body_mass_g", x ="species" , data = penguins)
```

Adding a palette to boxplots

```python
sns.boxplot(y ="body_mass_g", x ="species" , data = penguins, palette = "Set2")
```

Changing the boundary lines width

```python
sns.boxplot(y ="body_mass_g", x ="species" , data = penguins, linewidth = 3)
```

Changing the order of the boxplots

```python
sns.boxplot(y ="body_mass_g", x ="species" , data = penguins, linewidth = 3, order = ["Chinstrap" , "Adelie", "Gentoo"])
```

Displays boxplots for the males and females for all species

```python
sns.boxplot(y ="body_mass_g", x ="sex" , hue = "species", data = penguins)
```
